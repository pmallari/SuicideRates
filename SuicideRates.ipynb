```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm_notebook
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
plt.style.use('seaborn')
```

```python
df = pd.read_csv("data/who_suicide_statistics.csv")
```

```python
df.head()
```

```python
df = df[(df['year'] >= 1985) & (df['year'] < 2015)]
```

The structure tells us that the data is bucketed into groups based on Country, Year, Sex, and Age. For each demograph, we can find the suicide count and the population of each demograph.

We can improve the data by adding a continent for each country. We will be using the data provided in this link 'https://datahub.io/JohnSnowLabs/country-and-continent-codes-list#data'

```python
df2 = pd.read_csv("data/country_data.csv")
```

We'll create a dictionary for each country in the dataset and spot if there are any countries that wasn't listed or didn't match the continent dataset.

```python
country_to_continent = {}
not_found = []
for country in set(df['country'].values):
    match = df2[df2['Country_Name'].str.contains(country)]['Continent_Name']
    if len(match) > 0:
        country_to_continent[country] = df2[df2['Country_Name'].str.contains(country)]['Continent_Name'].iloc[0]
    else:
        not_found.append(country)

print("Countries not found in the country-continent dataset: \n")
print('\n'.join(not_found))
```

For the missing countries, we will fill it in manually.

```python
country_to_continent['Cabo Verde'] = 'Africa'
country_to_continent['Virgin Islands (USA)'] = 'North America'
country_to_continent['Venezuela (Bolivarian Republic of)'] = 'South America'
country_to_continent['Rodrigues'] = 'Asia'
country_to_continent['Kyrgyzstan'] = 'Asia'
country_to_continent['Saint Vincent and Grenadines'] = 'North America'
country_to_continent['TFYR Macedonia'] = 'Europe'
country_to_continent['Republic of Moldova'] = 'Europe'
country_to_continent['Iran (Islamic Rep of)'] = 'Asia'
country_to_continent['Occupied Palestinian Territory'] = 'Asia'
country_to_continent['Macau'] = 'Asia'
country_to_continent['Republic of Korea'] = 'Asia' 
country_to_continent['Falkland Islands (Malvinas)'] = 'South America'
country_to_continent['Hong Kong SAR'] = 'Asia'
```

```python
df['continent'] = [country_to_continent[country] for country in df['country']]
```

```python
df = df.reindex([df.columns[-1]] + list(df.columns[:-1]), axis = 1)
```

We will also add a percentage metric to show the percentage of the suicides_no to the population. If suicides_no is missing or the population is missing, we will add np.NaN.

```python
percentage = []
per_100k = []
for count, population in df[['suicides_no', 'population']].values:
    if (count == count) & (population == population):
        percentage.append(count/population)
        per_100k.append(int(count * (100000/population)))
    else:
        percentage.append(np.NaN)
        per_100k.append(np.NaN)
        
df['percentage'] = percentage
df['per_100k'] = per_100k
df['age'].replace("5-14 years", "05-14 years", inplace = True) 
```

One key thing to keep in mind is that the available data varies per year. More data is collected from recent years compared to way back.

```python
curr = df.dropna().groupby(['year']).count().reset_index()
plt.figure(figsize = (8,6))
plt.plot(curr['year'], curr['suicides_no'], '--o--', label = "Available Datapoints")
plt.plot(curr['year'], [curr['suicides_no'].mean()] * len(curr), '--', label = "Mean")
plt.title("Datapoints per year", fontsize = 18)
plt.xlabel("Year", fontsize = 14)
plt.ylabel("Datapoints", fontsize = 14)
plt.legend()
plt.show()
```

One analysis that we can do is to look at the global rate of suicide.

```python
curr = df.dropna().groupby(['year']).sum().reset_index()
curr['per_100k'] = curr['suicides_no'] * 100000 / curr['population']
plt.figure(figsize = (8,6))
plt.plot(curr['year'], curr['per_100k'], '--o--', label = "Suicide Rate")
plt.plot(curr['year'], [curr['per_100k'].mean()] * len(curr), '--', label = "Mean")
plt.title("Global Suicide Rate from 1985 to 2015", fontsize = 16)
plt.xlabel("Year", fontsize = 14)
plt.ylabel("Suicide Rate per 100,000", fontsize = 14)
plt.legend()
plt.show()
```

A noticeable insight here is that the **rate of suicide has been declining** over the past 30 years.

We can segment the data into different age groups to understand how the statistics is broken up.

```python
plt.figure(figsize=(6,6))
curr = df.groupby(['age']).sum().reset_index()
plt.bar(curr['age'].values, curr['suicides_no'].values * 100000 / curr['population'].values)
plt.title('Global Suicide Count per Age Group', fontsize = 14)
plt.xlabel('Age Group', fontsize = 14)
plt.ylabel('Suicide Count per 100,000', fontsize = 14)
plt.xticks(rotation = 45)
plt.show()
```

We can see that the group with the highest suicide count per 100,000 individuals are among the 75+ group. We can also see that the **suicide count increases as the age increases**.

```python
curr = df.groupby(['year', 'age']).sum().reset_index()
plt.figure(figsize = (8,6))
for age_group in curr['age'].unique():
    sub = curr[curr['age'] == age_group]
    sub['per_100k'] = sub['suicides_no'] * 100000 / sub['population']
    plt.plot(sub['year'], sub['per_100k'], '--o--', label = age_group)

plt.title('Suicide count from 1985 - 2015', fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Suicide Count per 100,000', fontsize = 14)
plt.legend()
plt.show()

plt.figure(figsize = (8,6))
for age_group in curr['age'].unique():
    sub = curr[curr['age'] == age_group]
    plt.plot(sub['year'], sub['suicides_no'], '--o--', label = age_group)

plt.title('Suicide count from 1985 - 2015', fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Total Suicide Count', fontsize = 14)
plt.legend()
plt.show()
```

Suicide rate may be decreasing over the past years however the total count of suicide has plateaued.

```python
curr = df[(df['suicides_no'] != np.NaN) & (df['population'] != np.NaN)]
curr = df.groupby(['year', 'sex']).sum().reset_index()
plt.figure(figsize = (8,6))
for sex in curr['sex'].unique():
    sub = curr[curr['sex'] == sex]
    sub['per_100k'] = sub['suicides_no'] * 100000 / sub['population']
    plt.plot(sub['year'], sub['per_100k'], '--o--', label = sex)

plt.title('Suicide count based on Gender', fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Suicide count per 100,000', fontsize = 14)
plt.ylim(bottom = 0)
plt.legend()
plt.show()

plt.figure(figsize = (8,6))
for sex in curr['sex'].unique():
    sub = curr[curr['sex'] == sex]
    plt.plot(sub['year'], sub['suicides_no'], '--o--', label = sex)

plt.title('Suicide count based on Gender', fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Total Suicide Count', fontsize = 14)
plt.ylim(bottom = 0)
plt.legend()
plt.show()
```

Another interesting trend that we notice is that the **suicide rate of men is around 3-4 times as that of women**. This trend is apparent both in the rate as well as the total count.

```python
curr = df.dropna().groupby(['year', 'continent']).sum().reset_index()
plt.figure(figsize = (8,6))
for continent in curr['continent'].unique():
    sub = curr[curr['continent'] == continent]
    sub['per_100k'] = sub['suicides_no'] * 100000 / sub['population']
    plt.plot(sub['year'], sub['per_100k'], '--o--', label = continent)

plt.title('Suicide count from 1985 - 2015', fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Suicide count per 100,000', fontsize = 14)
plt.legend()
plt.show()
```

We can see that the suicide rate is high specifically in **Europe, Asia, and Oceania**.

```python
curr = df.dropna().groupby(['year', 'age']).sum().reset_index()
plt.figure(figsize = (8,6))
for age in curr['age'].unique():
    sub = curr[curr['age'] == age]
    plt.plot(sub['year'], sub['suicides_no'], '--o--', label = age)

plt.title('Suicide count from 1985 - 2015 per age group', fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Suicide count', fontsize = 14)
plt.legend()
plt.show()
```
